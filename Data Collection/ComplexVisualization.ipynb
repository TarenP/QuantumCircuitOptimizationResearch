```python
def lstSub(list1, list2):
    #empty list
    sub_list = []

    #two lists are passed to zip 
    zip_object = zip(list1, list2)

    #loop to find diff of each element
    for list1_i, list2_i in zip_object:
        sub_list.append(list1_i - list2_i)
        
    return(sub_list)
```

```python
import pickle
#Load MLModel
filename='MLModels//PerthSimRegModel.sav'
reg =pickle.load(open(filename, 'rb'))

def check(x):
    return x and [x[0]]*len(x) == x
def Selector(costs, GCs, RTTimes, TTTimes):
    tqc3RT = float(RTTimes[3]) + float(TTTimes[3])
    tqc2RT = float(RTTimes[2]) + float(TTTimes[2])
    tqc1RT = float(RTTimes[1]) + float(TTTimes[1])
    tqc0RT = float(RTTimes[0]) + float(TTTimes[0])
    RTs = [tqc0RT, tqc1RT, tqc2RT, tqc3RT]


    RTs2 = [RTs[0], RTs[1], RTs[2], RTs[3]]
    GCs2 = [GCs[0], GCs[1], GCs[2], GCs[3]]
    costs2 = [costs[0], costs[1], costs[2], costs[3]]

    optLvls = [0, 1, 2, 3]
    worst = RTs.index(max(RTs))
    optLvls.pop(worst)
    RTs.pop(worst)
    GCs.pop(worst)
    costs.pop(worst)
    #all items are identical
    if check(costs):
        worst = RTs.index(max(RTs))
    else:
        worst = costs.index(max(costs))
    optLvls.pop(worst)
    RTs.pop(worst)
    GCs.pop(worst)
    costs.pop(worst)
    if check(GCs):
        worst = RTs.index(max(RTs))
    else:
        worst = GCs.index(max(GCs))
    optLvls.pop(worst)
    best = optLvls[0]
    # print("old best: ", best)
    #Check if any RTs are too close to be distinguished
    alternatives = []
    for i in range(len(RTs2)):
        if i != best:
            # print(i)
            # print((RTs2[i] - RTs2[best]) / RTs2[best] *100)
            if (RTs2[i] - RTs2[best]) / RTs2[best] *100 <= 30 and GCs2[i] < GCs2[best] and costs2[i] < costs2[best]:
                best = i
            elif (RTs2[i] - RTs2[best]) / RTs2[best] *100 <= 5 and GCs2[i] <= GCs2[best] and costs2[i] <= costs2[best]:
                alternatives.append(i)
    return best, alternatives

def sel(costs, GCs, TTTimes):
    
    RT3 = reg.predict([[costs[3], GCs[3], 3, TTTimes[3]]])

    RT2 = reg.predict([[costs[2], GCs[2], 2, TTTimes[2]]])

    RT1 = reg.predict([[costs[1], GCs[1], 1, TTTimes[1]]])

    RT0 = reg.predict([[costs[0], GCs[0], 0, TTTimes[0]]])
    

    #Get runtime of each optimization level for this qc
    tqc3RT = RT3[0] + TTTimes[3]
    tqc2RT = RT2[0] + TTTimes[2]
    tqc1RT = RT1[0] + TTTimes[1]
    tqc0RT = RT0[0] + TTTimes[0]
    RTs = [tqc0RT, tqc1RT, tqc2RT, tqc3RT]
    
    RTs2 = [RTs[0], RTs[1], RTs[2], RTs[3]]
    GCs2 = [GCs[0], GCs[1], GCs[2], GCs[3]]
    costs2 = [costs[0], costs[1], costs[2], costs[3]]
    # best = RTs.index(min(tqc0RT, tqc1RT, tqc2RT, tqc3RT))

    #Cascade choice
    #RT
    #Gatecount
    #Cost
    optLvls = [0, 1, 2, 3]
    worst = RTs.index(max(RTs))
    optLvls.pop(worst)
    RTs.pop(worst)
    GCs.pop(worst)
    costs.pop(worst)
    #all items are identical
    if check(costs):
        worst = RTs.index(max(RTs))
    else:
        worst = costs.index(max(costs))
    optLvls.pop(worst)
    RTs.pop(worst)
    GCs.pop(worst)
    costs.pop(worst)
    if check(GCs):
        worst = RTs.index(max(RTs))
    else:
        worst = GCs.index(max(GCs))
    optLvls.pop(worst)
    best = optLvls[0]
    
    #Check if any RTs are too close to be distinguished
    alternatives = []
    for i in range(len(RTs)):
        if i != best:
            if (RTs2[i] - RTs2[best]) / RTs2[best] *100 <= 30 and GCs2[i] < GCs2[best] and costs2[i] < costs2[best]:
                best = i
            elif (RTs2[i] - RTs2[best]) / RTs2[best] *100 <= 5 and GCs2[i] <= GCs2[best] and costs2[i] <= costs2[best]:
                alternatives.append(i)
    return best, alternatives
```

```python
def Accuracy(selBest, selAlts, qcBest, qcAlts):
    #Get Accuracy
    #simVal is the array of selected optimization levels by my Selector function
    #controlVal is the array of selected optimization levels by the control
    correct = 0
    # for i in range(len(selBest)):
    point = False
    if selBest == qcBest:
        correct += 1
        point = True
    for qcAlt in qcAlts:
        if selBest == qcAlt and point == False:
            correct += 1
            point = True
            break
    for alt in selAlts:
        if alt == qcAlts and point == False:
            correct += 1
            point = True
            break
    for alt in selAlts:
        for qcAlt in qcAlts:
            if alt == qcAlt and point == False:
                correct += 0.5
                point = True
                break

    # return 100* (correct/(len(selBest)))
    return 100* (correct)
```
